### **7.4 Gráficos, diagnóstico y vinculación**

### **Gráfico 1: Redes vs Fanatismo**

```python
# Preparo las columnas de texto (mapeo de IDs)

# Mapeo Compra de entradas
mapa_entradas = {
    i+1: val
    for i, val in enumerate(all_sheets["Compra entradas"].iloc[:,0])
}
df["Compra_entradas_texto"] = df["Compra entradas"].map(mapa_entradas)

# Mapeo Género
mapa_genero = {
    i+1: val
    for i, val in enumerate(all_sheets["Genero"].iloc[:,0])
}
df["Genero_texto"] = df["Genero"].map(mapa_genero)

# Mapeo País de residencia
mapa_pais = {
    i+1: val
    for i, val in enumerate(all_sheets["Pais Residencia"].iloc[:,0])
}
df["Pais_residencia_texto"] = df["Pais de residencia"].map(mapa_pais)

# Mapeo “Sigo a Taylor en redes”
mapa_redes = {
    i+1: val
    for i, val in enumerate(all_sheets["Red social"].iloc[:,0])
}
df["Sigo_redes_texto"] = df["Sigo a Taylor en redes"].map(mapa_redes)

# Mapeo “Compra vinilo”
mapa_vinilo = {
    i+1: val
    for i, val in enumerate(all_sheets["Vinilo"].iloc[:,0])
}
df["Compra_vinilo_texto"] = df["Compra vinilo"].map(mapa_vinilo)

# Mapeo “Nivel educativo alcanzado”
mapa_edu = {
    i+1: val
    for i, val in enumerate(all_sheets["Nivel Educativo"].iloc[:,0])
}
df["Nivel_educativo_texto"] = df["Nivel educativo alcanzado"].map(mapa_edu)
```

```python
plt.figure(figsize=(9, 5))
sns.boxplot(
    x="Compra_entradas_texto",
    y="Nivel Fanatismo",
    hue="Sigo_redes_texto",
    data=df,
    palette="Set3",
    width=0.6,
    fliersize=3
)
plt.title("Fanatismo según compra de entradas y seguimiento en redes")
plt.xlabel("Compra de entradas")
plt.ylabel("Nivel de fanatismo (1–10)")
plt.grid(axis='y', linestyle='--', alpha=0.4)
plt.xticks(rotation=30)
plt.legend(title="Sigo en redes", bbox_to_anchor=(1.02, 1))
plt.tight_layout()
plt.show()
```

### **Diagnóstico:**

- La mediana de fanatismo para quienes compraron entradas y siguen a Taylor en redes es de 10, frente a 8 para los que no la siguen.
- En todas las categorías de compra de entradas (asistieron, quisieron asistir, no pudieron, etc.), el fanatismo medio y el rango intercuartílico son sistemáticamente mayores para el grupo “Sí” en redes.

### **Interpretación:**

Los resultados validan la hipótesis H1 (*Quienes siguen a Taylor en redes (Sí) tienen un fanatismo promedio mayor que quienes no la siguen (No).*): la presencia digital—medida como el seguimiento en redes—se asocia a un nivel de fanatismo significativamente mayor, incluso controlando por la experiencia de asistir a un concierto.

### **Gráfico 3: Edad vs Fanatismo**

```python
# 1) Creo rangos de edad para simplificar la lectura
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Defino los bins y etiquetas
bins = [0, 25, 35, 45, 100]
labels = ["<25", "25–35", "36–45", "46+"]
df["Edad_grupo"] = pd.cut(df["Edad"], bins=bins, labels=labels)

# 2) Calculo fanatismo promedio por grupo de edad y género
fan_prom_edad = (
    df
    .groupby(["Edad_grupo", "Genero_texto"])["Nivel Fanatismo"]
    .mean()
    .reset_index()
)

# 3) Lineplot para mostrar la tendencia por género
plt.figure(figsize=(8, 5))
sns.lineplot(
    x="Edad_grupo",
    y="Nivel Fanatismo",
    hue="Genero_texto",
    data=fan_prom_edad,
    marker="o",
    linewidth=2,
    palette="tab10"
)
plt.title("Fanatismo promedio por grupo de edad y género")
plt.xlabel("Grupo de edad")
plt.ylabel("Fanatismo promedio (1–10)")
plt.grid(True, linestyle="--", alpha=0.4)
plt.legend(title="Género", bbox_to_anchor=(1.02, 1))
plt.tight_layout()
plt.show()
```

### **Diagnóstico:**
- El fanatismo medio desciende de 6.1 en el grupo “<25” a 5.3 en “36–45”, con un ligero repunte a 5.5 en “46+” para el género masculino.
- El subgrupo “Prefiero no decirlo” presenta un pico de 7.0 en “25–35”, posiblemente por un menor tamaño de muestra pero alto compromiso.

### **Interpretación:**
Estos hallazgos confirman la hipótesis H5 (*Quienes conocen los Easter Eggs (Sí) tienen, en promedio, niveles de fanatismo y asistencia a conciertos superiores a los que no los conocen (No).*): los seguidores más jóvenes exhiben un mayor compromiso emocional (fanatismo) que los de mediana edad. Las variaciones por género añaden riqueza al análisis, indicando patrones específicos que podrían explorarse en etapas posteriores.


###**7.5 Creación de variables derivadas y su análisis**
En esta sección generamos columnas nuevas para facilitar el análisis:

1. **Antigüedad en el fandom**:  
   Se calcula como `2025 − Año en que te hiciste fan`, para medir cuántos años lleva cada encuestado siendo fan.

2. **Categoría de fanatismo (`Fanatismo_cat`)**:  
   Convertimos el puntaje numérico de “Nivel de fanatismo” (1–5) en tres grupos:
   - 1–2 → “Bajo”  
   - 3–4 → “Medio”  
   - 5   → “Alto”

3. **Compró_Entradas**:  
   Variable booleana que indica si el encuestado asistió a al menos un concierto (`Cantidad de asistencia a conciertos > 0`).

4. **Total_Compras_Merch**:  
   Suma de las columnas `Compra merch oficial` y `Compra vinilo`, para reflejar el volumen total de productos comprados.

Estos cuatro indicadores aportan información clave sobre tiempo como fan, grado de compromiso, participación en conciertos y hábitos de compra.

####**Creación de variable 'Antigüedad_fandom'**

```python
# Crear 'Antigüedad_fandom'
df['Antigüedad_fandom'] = 2025 - df['Anio en que te hiciste fan']

# Mostrar las primeras filas para confirmar que la nueva columna se creó correctamente
df[['Anio en que te hiciste fan', 'Antigüedad_fandom']].head()
```

####**Gráfico: Histograma de Antigüedad en el fandom**

```python
# Teóricamente, el máximo posible es 2025 - 2006 = 19 años.
max_teorico = 2025 - 2006  # = 19

# Verificamos el máximo que tenemos en el DataFrame
max_observado = df['Antigüedad_fandom'].max()
print(f"Mínimo observado: {df['Antigüedad_fandom'].min()}")
print(f"Máximo observado (antes de filtrar): {max_observado}")

# Filtrar antigüedades fuera del rango [0, max_teorico]
df_filtrado = df[(df['Antigüedad_fandom'] >= 0) & (df['Antigüedad_fandom'] <= max_teorico)]

# Recalculamos el máximo observado tras el filtrado
max_filtrado = df_filtrado['Antigüedad_fandom'].max()
print(f"Máximo observado tras filtrar a {max_teorico}: {max_filtrado}")


# Graficar histograma usando el rango real [0, max_filtrado]
bins = range(0, max_filtrado + 2)  # crea un bin para cada entero de 0 a max_filtrado

plt.figure(figsize=(6, 4))
plt.hist(
    df_filtrado['Antigüedad_fandom'],
    bins=bins,
    edgecolor='k',
    align='left'
)
plt.title('Distribución de Antigüedad en el fandom (0–19 años)')
plt.xlabel('Años como fan')
plt.ylabel('Frecuencia')
plt.xticks(bins)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

####**Creación de variable 'Fanatismo_cat'**

##**8. Método de selección de características (feature selection)**

Para reducir la dimensionalidad antes de entrenar nuestros modelos, usaremos **SelectKBest** con la función de puntuación χ² (chi‐cuadrado). Este método es adecuado para nuestro problema de clasificación binaria (“Compró_Entradas”) porque:

- Trabaja sobre variables numéricas o categóricas discretizadas.  
- Nos permite elegir las **k** variables más relevantes según su dependencia estadística con la etiqueta binaria.

### Pasos generales

1. **Preparar datos**  
   - Definir `X` con todas las características relevantes (edad, antigüedad, fanatismo, país, nivel educativo, etc.)  
   - Definir `y = Compró_Entradas` (True/False).
   - Convertir variables categóricas en dummies (one‐hot encoding).

2. **Aplicar SelectKBest(chi2)**  
   - Seleccionar las top k características (p. ej. k = 8).  
   - Mostrar las puntuaciones y quedarnos con las k mejores.

3. **Obtener el subconjunto de variables reducidas**  
   - Conservar solo esas k columnas para usarlas en el pipeline de modelado.

```python
# 1. Preparar datos para feature selection

from sklearn.feature_selection import SelectKBest, chi2

# Definir X y y
y = df['Compró_Entradas'].astype(int)  # Convertir True/False a 1/0

# Seleccionar las columnas que queremos considerar como features:
features_num = ['Edad', 'Antigüedad_fandom', 'Total_Compras_Merch', 'Nivel Fanatismo']
features_cat = ['Genero', 'Pais de residencia', 'Nivel educativo alcanzado', 'Ocupacion']

# Crear DataFrame con variables numéricas
X_num = df[features_num].copy()

# One‐hot encoding de variables categóricas
X_cat = pd.get_dummies(df[features_cat], drop_first=True)

# Unir numéricas y dummies
X = pd.concat([X_num, X_cat], axis=1)

# Asegurarnos de que todos los valores sean no negativos (requisito para chi2)
X = X.clip(lower=0)


# 2. Aplicar SelectKBest con chi2

# Seleccionamos k variables (por ejemplo, k = 8)
k = 8
selector = SelectKBest(score_func=chi2, k=k)

# Ajustamos selector a X (ya recortado) y y
selector.fit(X, y)

# Obtener puntajes χ² y p-values
scores = selector.scores_
pvalues = selector.pvalues_
features = X.columns

# Crear DataFrame con resultados ordenados
df_scores = pd.DataFrame({
    'feature': features,
    'chi2_score': scores,
    'p_value': pvalues
}).sort_values(by='chi2_score', ascending=False)

print("Top 10 características según χ² (orden descendente):")
print(df_scores.head(10), "\n")

# Obtener nombres de las k características seleccionadas
mask = selector.get_support()  # array booleana de longitud n_features
selected_features = features[mask]
print(f"Características seleccionadas (k={k}):")
print(list(selected_features))

# 3. Crear nuevo DataFrame solo con las k características
X_reducido = X[selected_features].copy()

# Verificamos que X_reducido tenga solo k columnas
print(f"\nDimensión de X original: {X.shape}")
print(f"Dimensión de X reducido: {X_reducido.shape}")
```

<a id="#9-elegir-algoritmo"></a>
##**9. Elegir y entrenar un algoritmo de clasificación**
###**Entrenamiento de un clasificador para “Compró_Entradas”**

En este bloque vamos a utilizar las 8 variables seleccionadas con SelectKBest (χ²) para entrenar
un `RandomForestClassifier` que prediga si un fan compró entradas (etiqueta binaria).
Primero dividimos los datos en entrenamiento (70 %) y validación (30 %), luego ajustamos el
modelo y calculamos métricas básicas (accuracy, precisión, recall, F1 y matriz de confusión).

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

# 1. Definir X_reducido y y para la clasificación
#    (ya tienes X_reducido con las 8 features seleccionadas)
X_clas = X_reducido.copy()    # X_reducido proviene del paso ii
y_clas = df['Compró_Entradas'].astype(int)  # 1 si asistió a ≥1 concierto, 0 si no

# 2. Separar en conjuntos de entrenamiento y validación (70 % / 30 %)
X_train, X_val, y_train, y_val = train_test_split(
    X_clas, y_clas,
    test_size=0.30,
    random_state=42,
    stratify=y_clas
)

# 3. Instanciar el clasificador (RandomForestClassifier)
clf = RandomForestClassifier(
    n_estimators=100,     # número de árboles
    max_depth=None,       # sin límite de profundidad
    min_samples_leaf=1,   # hoja mínima de 1 muestra
    random_state=42
)

# 4. Entrenar el modelo
clf.fit(X_train, y_train)

# 5. Predecir en el conjunto de validación
y_pred = clf.predict(X_val)

# 6. Calcular métricas básicas para validar el modelo
acc  = accuracy_score(y_val, y_pred)
prec = precision_score(y_val, y_pred)
rec  = recall_score(y_val, y_pred)
f1   = f1_score(y_val, y_pred)
cm   = confusion_matrix(y_val, y_pred)

print(f"Accuracy:  {acc:.3f}")
print(f"Precision: {prec:.3f}")
print(f"Recall:    {rec:.3f}")
print(f"F1‐score:  {f1:.3f}")
print("\nMatriz de confusión:")
print(cm)
```
